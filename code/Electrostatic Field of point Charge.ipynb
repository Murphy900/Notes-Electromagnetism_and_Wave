```python
#Dieletric costant and unitary charge - - - - - -- - - -

DIEL = 8.854187822*10**(-12) 
Q =  1.6*10**(-19) #charge of a proton

# - - - -- - - - - - -- - - - - - - - - - - - - - - - -

q1 = Q
q2 = -3*Q


#Electrostatic field of a point charge 

def Ex (x,y,q,x0,y0):
    return q/(4.0*pi*DIEL)*(x-x0)/((x-x0)**2+(y-y0)**2)**(1.5)

def Ey (x,y,q,x0,y0):
    return q/(4*pi*DIEL)*(y-y0)/((x-x0)**2+(y-y0)**2)**(1.5)


# Total electrostatic field of two point charge at distance d on the x-axis 

Exx =  lambda x,y: Ex(x,y,q1,-3,0) + Ex(x,y,q2,3,0)
Eyy =  lambda x,y: Ey(x,y,q1,-3,0) + Ey(x,y,q2,3,0)

#plotting the normalized vector field (the problem with normalization is that we loss information about the strentgh of the field 

x,y = var('x y')
P = plot_vector_field((Exx(x,y)/sqrt(Exx(x,y)**2+Eyy(x,y)**2),Eyy(x,y)/sqrt(Exx(x,y)**2+Eyy(x,y)**2)),(x,-5,5), (y,-5,5))
C1 = point((-3,0),size = 150) + text('q',(-2.7,-0.5),fontsize = 20)
C2 = point((3,0),size = 150) + text('-3q',(3.2,0.5),fontsize = 20)
S = P+C1+C2
S.axes_labels([r'$E_x$',r'$E_y$'])
S.show(title = 'Electrostatic Field of a Point charge q ad -3q', fontsize = 15)
```
